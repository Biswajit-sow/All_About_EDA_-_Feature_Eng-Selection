```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=True,cmap='viridis') #heatmap is a Seaborn function that plots a rectangular grid where each cell is colored based on its value.Here, the heatmap will show where missing values exist in the DataFrame.
#yticklabels=False This hides the row labels (Y-axis labels) of the heatmap. Useful when you have many rows and want a cleaner plot.
# cbar=False This removes the color bar (the legend on the side showing value-to-color mapping).
#cmap='viridis' cmap stands for colormap – it defines the colors used in the heatmap. 'viridis' is a color gradient from dark purple → yellow.Seaborn uses this to color the True/False matrix, so True (missing) might appear bright, False (not missing) darker.
```

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'}) #reset_index help to convert to dataframe
```

```python
ratings
```

Observation
1. when Rating is between 4.5 to 4.9 ------ Exellent 
2. when Rating is between 4.0 to 4.9------very good 
3.when Rating is between 3.5 to 3.9------good 
4.when Rating is between 2.5 to 3.4------average 
5.when Rating is between 1.8 to 2,4------very good 
6.when Rating is between 0.0----------people not rated

```python
ratings.head()
```

```python
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings) #data=ratings This specifies the DataFrame (ratings) from which the data is taken.
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=ratings) #data=ratings This specifies the DataFrame (ratings) from which the data is taken.
# Hue=Seaborn automatically picks colors from a palette and assigns them in the order the values appear, not based on specific text like 'Red', 'Green', etc.
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=ratings,palette=['white','red','orange','yellow','green','Darkgreen']) #data=ratings This specifies the DataFrame (ratings) from which the data is taken.
```

Observation:
1. Not rated(0.0)  count is very high which is white 
2. Maximum  number of rating are between  2.5 to 3.4 

How to handle if some rating are missing 

suppose a person has rated but you have some missing value now -then you use the maximum no of rating average from 2.5 to 3.4

```python
## Count plot 
# how frequently it is basically prsent in the data frame 
sns.countplot(x="Rating color",data=ratings,palette=['white','red','orange','yellow','green','Darkgreen'])
```
